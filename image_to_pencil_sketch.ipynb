**IMPORTING REQUIRED LIBRARY**

```python
from google.colab.patches import cv2_imshow
import cv2
```

**Display the image**

```python
image=cv2.imread("/content/Ganesha.jpg")
cv2_imshow(image)
```

**We can also resize the image from original form**

**Displaying the image in gray**

```python
img_gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
cv2_imshow(img_gray)
```

**displaying the image in inverted form**

```python
img_inverted=cv2.bitwise_not(img_gray)
cv2_imshow(img_inverted)
```

**Displaying the image in blur**

```python
img_blur=cv2.GaussianBlur(img_inverted,(21,21),0)
cv2_imshow(img_blur)
```

**Turning the image into sketch form**

```python
img_smooth=255-img_blur
img_sketch=cv2.divide(img_gray,img_smooth,scale=156.0)
cv2_imshow(img_sketch)
```

**At last we see original image and pencil sketch image**

```python
cv2_imshow(image)
cv2_imshow(img_sketch)
```
